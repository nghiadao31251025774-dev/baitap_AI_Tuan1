```python
import folium
from folium.plugins import HeatMap
from geopy.distance import geodesic
import numpy as np

# 1. Khởi tạo bản đồ tại UEH (Bài 23.1)
ueh_location = [10.7801, 106.6953]
m = folium.Map(location=ueh_location, zoom_start=14)
# 2. Thêm các địa điểm lân cận
locations = [
    {"name": "Bệnh viện", "coord": [10.7815, 106.6935]},
    {"name": "TTTM", "coord": [10.7818, 106.6995]},
    {"name": "Trạm xe buýt", "coord": [10.7800, 106.6940]}
]
for loc in locations:
    folium.Marker(loc['coord'], popup=loc['name']).add_to(m)
# 3. Vẽ vùng phục vụ 3km, 5km, 10km (Bài 23.5)
for radius in [3000, 5000, 10000]:
    folium.Circle(ueh_location, radius=radius, color='blue', fill=True, opacity=0.1).add_to(m)
# 4. Giả lập dữ liệu bản đồ nhiệt (Bài 23.3)
heat_data = [[10.78 + np.random.normal(0, 0.01), 106.69 + np.random.normal(0, 0.01), 1] for _ in range(100)]
HeatMap(heat_data).add_to(m)

# Dòng này để hiển thị bản đồ ra màn hình
m

```

```python
import osmnx as ox
import networkx as nx

# Tải mạng lưới đường bộ Quận 1 (Bài 23.6)
place_name = "District 1, Ho Chi Minh City, Vietnam"
G = ox.graph_from_place(place_name, network_type='drive')
# Lấy 2 điểm ngẫu nhiên để tìm đường (Bài 23.7)
nodes = list(G.nodes())
orig, dest = nodes[0], nodes[100]
# So sánh Dijkstra và A*
path_dijkstra = nx.shortest_path(G, orig, dest, weight='length')
path_astar = nx.astar_path(G, orig, dest, weight='length')
# Vẽ kết quả (Dùng làm hình ảnh cho báo cáo/poster)
fig, ax = ox.plot_graph_route(G, path_dijkstra, route_color='r', route_linewidth=4)
```

```python
import folium
from folium.plugins import HeatMap
import numpy as np
from sklearn.cluster import KMeans

# 1. Khởi tạo bản đồ
ueh_location = [10.7801, 106.6953]
m = folium.Map(location=ueh_location, zoom_start=14)
# 2. Giả lập dữ liệu khách hàng (Bài 23.3 & 23.9)
heat_data = [[10.78 + np.random.normal(0, 0.01), 106.69 + np.random.normal(0, 0.01)] for _ in range(100)]
HeatMap(heat_data).add_to(m)
# 3. AI Phân cụm tìm vị trí kho (Bài 23.9)
kmeans = KMeans(n_clusters=3, n_init=10).fit(heat_data)
for center in kmeans.cluster_centers_:
    folium.Marker(center.tolist(), icon=folium.Icon(color='red', icon='info-sign')).add_to(m)

# 4. HIỂN THỊ BẢN ĐỒ
m

```

```python
import time
from IPython.display import display

#BÀI 23.14: Mô phỏng hệ thống điều phối xe thời gian thực
def simulate_dispatching():
    print("--- Đang mô phỏng điều phối xe (Bài 23.14) ---")
    # Giả lập 1 xe di chuyển từ UEH đến Bitexco trong 5 bước thời gian
    start_point = np.array([10.7801, 106.6953]) # UEH
    end_point = np.array([10.7767, 106.7009])   # Bitexco
    # Tạo lộ trình di chuyển (chia nhỏ quãng đường)
    route = np.linspace(start_point, end_point, 5)
    m14 = folium.Map(location=ueh_location, zoom_start=15)
    for i, pos in enumerate(route):
        # Trong thực tế sẽ cập nhật tọa độ này lên Dashboard
        folium.Marker(
            location=pos.tolist(),
            icon=folium.Icon(color='blue', icon='car', prefix='fa'),
            popup=f"Xe đang ở vị trí bước {i+1}"
        ).add_to(m14)
        print(f"Bước {i+1}: Xe đang ở tọa độ {pos}")
    m14.save("bai23_14_mo_phong.html")
    print("-> Đã lưu bản đồ mô phỏng xe di chuyển.")
simulate_dispatching()
# BÀI 23.15: Ứng dụng AI đề xuất: Tối ưu trạm sạc xe điện
def ai_application_23_15():
    print("\n--- Ứng dụng AI đề xuất (Bài 23.15) ---")
    # Bước 1: Giả lập 50 điểm đỗ xe trong thành phố (Nhu cầu sạc)
    parking_locations = np.array(ueh_location) + np.random.uniform(-0.03, 0.03, (50, 2))
    # Bước 2: Dùng K-Means AI để tìm 5 vị trí đặt trạm sạc tối ưu nhất
    model = KMeans(n_clusters=5, n_init=10)
    model.fit(parking_locations)
    charging_stations = model.cluster_centers_
    # Bước 3: Trực quan hóa
    m15 = folium.Map(location=ueh_location, zoom_start=13)
    # Vẽ các điểm đỗ xe (màu xám)
    for p in parking_locations:
        folium.CircleMarker(p, radius=2, color='gray', opacity=0.5).add_to(m15)
    # Vẽ trạm sạc đề xuất (màu xanh lá - biểu tượng sét)
    for i, station in enumerate(charging_stations):
        folium.Marker(
            location=station.tolist(),
            popup=f"Trạm sạc AI đề xuất #{i+1}",
            icon=folium.Icon(color='green', icon='bolt', prefix='fa')
        ).add_to(m15)
    m15.save("bai23_15_ai_app.html")
    print("-> Đã hoàn thành ứng dụng AI tối ưu trạm sạc. Lưu tại bai23_15_ai_app.html")
ai_application_23_15()
```
