```python
#23.2
from geopy.geocoders import Nominatim
from geopy.distance import geodesic
import folium

# 1. Khởi tạo Geocoder (User_agent có thể đặt tên bất kỳ)
geolocator = Nominatim(user_agent="geo_exercise")

# 2. Danh sách 5 địa chỉ mẫu (bạn có thể mở rộng lên 10 theo yêu cầu)
addresses = [
    "Dinh Độc Lập, TP.HCM",
    "Bưu điện Trung tâm Sài Gòn",
    "Chợ Bến Thành, TP.HCM",
    "Nhà hát Thành phố Hồ Chí Minh",
    "Hồ Con Rùa, Quận 3, TP.HCM"
]

locations_data = []

# 3. Chuyển địa chỉ sang tọa độ và lưu trữ
print("Đang lấy tọa độ...")
for addr in addresses:
    loc = geolocator.geocode(addr)
    if loc:
        locations_data.append({
            "name": addr,
            "lat": loc.latitude,
            "lon": loc.longitude
        })

# 4. Tính khoảng cách từ điểm đầu tiên đến các điểm còn lại
start_point = (locations_data[0]['lat'], locations_data[0]['lon'])
print(f"\nKhoảng cách từ: {locations_data[0]['name']}")

for i in range(1, len(locations_data)):
    end_point = (locations_data[i]['lat'], locations_data[i]['lon'])
    dist = geodesic(start_point, end_point).km
    print(f"- Đến {locations_data[i]['name']}: {dist:.2f} km")

# 5. Trực quan hóa lên bản đồ Folium
m = folium.Map(location=start_point, zoom_start=14)

for loc in locations_data:
    folium.Marker(
        [loc['lat'], loc['lon']],
        popup=loc['name'],
        icon=folium.Icon(color='blue', icon='info-sign')
    ).add_to(m)

m.save('khoang_cach_geopy.html')
```

```python
#23.8
import folium
import numpy as np

# 1. Tạo tọa độ trung tâm (Ví dụ: Quận 1, TP.HCM)
center = [10.7769, 106.7009]
m = folium.Map(location=center, zoom_start=15)
# 2. Giả lập dữ liệu 5 xe và 3 khách hàng (Tọa độ ngẫu nhiên quanh trung tâm)
drivers = {
    'Xe A': [10.7780, 106.7000],
    'Xe B': [10.7750, 106.7050],
    'Xe C': [10.7800, 106.7020],
    'Xe D': [10.7720, 106.6980],
    'Xe E': [10.7760, 106.7080]
}
customers = {
    'Khách 1': [10.7775, 106.7015],
    'Khách 2': [10.7730, 106.7030]
}

# 3. Thuật toán Heuristic: Ghép khách với xe gần nhất (Euclidean distance)
def get_distance(p1, p2):
    return np.sqrt((p1[0]-p2[0])**2 + (p1[1]-p2[1])**2)
assignments = {}
for c_name, c_coords in customers.items():
    # Tìm xe có khoảng cách nhỏ nhất tới khách
    closest_driver = min(drivers.keys(), key=lambda d: get_distance(c_coords, drivers[d]))
    assignments[c_name] = closest_driver
    # Vẽ đường nối giữa khách và xe được chọn
    folium.PolyLine([c_coords, drivers[closest_driver]], color="blue", weight=2, dash_array='5, 5').add_to(m)
# 4. Hiển thị lên bản đồ
for d_name, d_coords in drivers.items():
    folium.Marker(d_coords, popup=d_name, icon=folium.Icon(color='green', icon='car', prefix='fa')).add_to(m)
for c_name, c_coords in customers.items():
    folium.Marker(c_coords, popup=f"{c_name} -> {assignments[c_name]}", icon=folium.Icon(color='red', icon='user', prefix='fa')).add_to(m)

# Hiển thị bản đồ
m
```

```python
#23.9
import folium
from sklearn.cluster import KMeans
import numpy as np

# 1. Giả lập dữ liệu tọa độ của 20 khách hàng (phân bố ngẫu nhiên)
np.random.seed(42)
customer_coords = np.random.normal(loc=[10.776, 106.700], scale=[0.01, 0.01], size=(20, 2))
# 2. Áp dụng thuật toán K-Means để phân thành 3 cụm (3 kho hàng dự kiến)
# Mỗi cụm sẽ có một "Centroid" - đây chính là vị trí đặt kho hàng tối ưu
n_clusters = 3
kmeans = KMeans(n_clusters=n_clusters, random_state=0).fit(customer_coords)
centers = kmeans.cluster_centers_ # Tọa độ các trạm trung chuyển
labels = kmeans.labels_ # Nhãn cụm của từng khách hàng
# 3. Trực quan hóa lên bản đồ
m = folium.Map(location=[10.776, 106.700], zoom_start=14)
# Vẽ khách hàng (Marker nhỏ)
colors = ['red', 'blue', 'green']
for i in range(len(customer_coords)):
    folium.CircleMarker(
        location=customer_coords[i],
        radius=5,
        color=colors[labels[i]],
        fill=True,
        popup=f"Khách hàng cụm {labels[i]}"
    ).add_to(m)
# Vẽ trạm trung chuyển (Marker lớn đặc biệt)
for i, center in enumerate(centers):
    folium.Marker(
        location=center,
        popup=f"TRẠM TRUNG CHUYỂN TỐI ƯU {i+1}",
        icon=folium.Icon(color='black', icon='university', prefix='fa')
    ).add_to(m)

# Hiển thị bản đồ
m
```

```python
#23.11
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
import folium

# 1. Giả lập dữ liệu lịch sử nhu cầu
data = {
    'area_id': np.random.randint(0, 5, 100), # 5 khu vực
    'hour': np.random.randint(0, 24, 100),   # Từ 0h - 23h
    'is_weekend': np.random.randint(0, 2, 100), # 0: ngày thường, 1: cuối tuần
    'demand': np.random.randint(10, 100, 100)  # Số lượng đơn hàng
}
df = pd.DataFrame(data)
# 2. Huấn luyện mô hình AI (Random Forest)
X = df[['area_id', 'hour', 'is_weekend']]
y = df['demand']
model = RandomForestRegressor(n_estimators=100)
model.fit(X, y)
# 3. Dự báo nhu cầu cho khung giờ cao điểm (ví dụ: 18h tối nay)
test_hour = 18
is_wknd = 0 # Thứ Hai
areas = {
    'Quận 1': {'id': 0, 'coords': [10.776, 106.700]},
    'Quận 3': {'id': 1, 'coords': [10.782, 106.688]},
    'Quận 5': {'id': 2, 'coords': [10.755, 106.665]},
    'Quận 7': {'id': 3, 'coords': [10.732, 106.721]},
    'Quận 10': {'id': 4, 'coords': [10.771, 106.666]}
}
m = folium.Map(location=[10.776, 106.700], zoom_start=13)
print(f"--- Dự báo nhu cầu lúc {test_hour}h ---")
for name, info in areas.items():
    # Dự báo bằng model
    pred = model.predict([[info['id'], test_hour, is_wknd]])[0]
    # Trực quan hóa: Nhu cầu cao -> Vòng tròn to, màu đỏ; Thấp -> Nhỏ, màu xanh
    color = 'red' if pred > 60 else 'orange' if pred > 40 else 'blue'
    folium.CircleMarker(
        location=info['coords'],
        radius=pred/3, # Tỉ lệ hóa kích thước
        color=color,
        fill=True,
        popup=f"{name}: Dự báo {int(pred)} đơn"
    ).add_to(m)
    print(f"{name}: {int(pred)} đơn")

# 4. Hiển thị bản đồ
m
```

```python
#23.12
import folium
from scipy.spatial import distance

# 1. Định nghĩa vị trí 2 kho hàng (Depots) và 10 khách hàng
depots = [[10.776, 106.700], [10.755, 106.665]]
customers = [
    [10.780, 106.695], [10.772, 106.705], [10.785, 106.710],
    [10.760, 106.670], [10.750, 106.660], [10.745, 106.680],
    [10.790, 106.690], [10.770, 106.685], [10.765, 106.715], [10.758, 106.698]
]
m = folium.Map(location=[10.77, 106.68], zoom_start=13)
# 2. Thuật toán Heuristic đơn giản: Gán khách hàng cho kho gần nhất
routes = {0: [depots[0]], 1: [depots[1]]}
for cust in customers:
    # Tính khoảng cách từ khách hàng đến từng kho
    dist_to_depot0 = distance.euclidean(cust, depots[0])
    dist_to_depot1 = distance.euclidean(cust, depots[1])
    if dist_to_depot0 < dist_to_depot1:
        routes[0].append(cust)
    else:
        routes[1].append(cust)
# 3. Vẽ lên bản đồ
colors = ['blue', 'red']
for i, depot in enumerate(depots):
    # Vẽ kho hàng
    folium.Marker(depot, popup=f"Kho {i+1}", icon=folium.Icon(color='black', icon='home')).add_to(m)
    # Vẽ các khách hàng và đường đi của xe từ kho đó
    route = routes[i]
    for j in range(1, len(route)):
        folium.Marker(route[j], icon=folium.Icon(color=colors[i], icon='user')).add_to(m)
        # Vẽ đường nối (tuyến giao hàng)
        folium.PolyLine([route[j-1], route[j]], color=colors[i], weight=2.5, opacity=0.8).add_to(m)
    # Quay về kho sau khi giao xong
    folium.PolyLine([route[-1], route[0]], color=colors[i], weight=2.5, dash_array='5, 5').add_to(m)

m
```

```python
#23.14
import folium
from folium.plugins import AntPath
import osmnx as ox
import networkx as nx

# 1. Lấy mạng lưới giao thông khu vực Quận 1
G = ox.graph_from_place("District 1, Ho Chi Minh City, Vietnam", network_type='drive')
# 2. Xác định điểm đi và điểm đến (Chọn 2 nút ngẫu nhiên trên đồ thị)
nodes = list(G.nodes())
start_node = nodes[10]
end_node = nodes[100]
# 3. Thuật toán tìm đường ngắn nhất (Lộ trình của xe)
route = nx.shortest_path(G, start_node, end_node, weight='length')
# 4. Chuyển đổi các nút trong lộ trình thành tọa độ (Lat, Lon)
route_coords = []
for node in route:
    point = G.nodes[node]
    route_coords.append([point['y'], point['x']])
# 5. Tạo bản đồ mô phỏng
m = folium.Map(location=route_coords[0], zoom_start=15)
# 6. Sử dụng AntPath để tạo hiệu ứng xe đang di chuyển trên lộ trình
AntPath(
    locations=route_coords,
    dash_array=[10, 20],
    delay=1000, # Tốc độ di chuyển (ms)
    color='blue',
    pulse_color='white',
    paused=False,
    reverse=False,
    tooltip='Xe đang giao hàng...'
).add_to(m)
# Thêm Marker cho điểm đầu và điểm cuối
folium.Marker(route_coords[0], popup='Vị trí xuất phát', icon=folium.Icon(color='green')).add_to(m)
folium.Marker(route_coords[-1], popup='Điểm đến', icon=folium.Icon(color='red', icon='flag')).add_to(m)
# 7. Hiển thị mô phỏng
m
```

```python
#23,15
import folium
from folium.plugins import HeatMap
import numpy as np

# 1. Khởi tạo bản đồ Quận 1
m = folium.Map(location=[10.776, 106.700], zoom_start=15)
# 2. Giả lập dữ liệu đối thủ (Ví dụ: các thương hiệu lớn hiện có)
competitors = np.random.normal(loc=[10.776, 106.700], scale=[0.005, 0.005], size=(15, 2))
# 3. Giả lập dữ liệu khách hàng mục tiêu (ví dụ: các trường học, văn phòng)
customers = np.random.normal(loc=[10.778, 106.702], scale=[0.007, 0.007], size=(50, 2))
# 4. Vẽ Bản đồ nhiệt của KHÁCH HÀNG (Vùng tiềm năng)
HeatMap(customers, radius=20, blur=15, name="Mật độ khách hàng").add_to(m)
# 5. Đánh dấu các ĐỐI THỦ (Vùng rủi ro)
for coord in competitors:
    folium.Circle(
        location=coord,
        radius=200, # Bán kính ảnh hưởng của đối thủ
        color='red',
        fill=True,
        fill_opacity=0.3,
        popup="Đối thủ cạnh tranh"
    ).add_to(m)
# 6. Đề xuất vị trí tối ưu bằng AI (Tìm điểm đông khách nhưng ngoài bán kính đối thủ)
# Ở đây ta giả lập một vị trí tiềm năng nhất
optimal_location = [10.782, 106.698]
folium.Marker(
    location=optimal_location,
    popup="<b>VỊ TRÍ ĐỀ XUẤT MỞ CỬA HÀNG</b>",
    icon=folium.Icon(color='gold', icon='star', prefix='fa')
).add_to(m)
folium.LayerControl().add_to(m)
m
```
